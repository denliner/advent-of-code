```python
import numpy as np
from collections import defaultdict
from collections import deque

def generate_list_intructions(name):
    with open(name) as f:
        input_=f.read().strip().splitlines()
        return [x.split(" ") for x in input_]
#


def doInstruction(instruction,variableDict,inputVal):
    if(instruction[0]=="inp"):
        variableDict[instruction[1]]=int(inputVal.pop(0))
    else:
        if(instruction[2].isalpha()):
            second=variableDict[instruction[2]]
        else:
            second=int(instruction[2])
      #  print(instruction[1],variableDict[instruction[1]],second)
        if(instruction[0]=="add"):
     #       print("second",second)
            variableDict[instruction[1]]+=second
        elif(instruction[0]=="mul"):
            variableDict[instruction[1]]*=second

        elif(instruction[0]=="div"):
            variableDict[instruction[1]]=int(variableDict[instruction[1]]/second)
        elif(instruction[0]=="mod"):
            variableDict[instruction[1]]=variableDict[instruction[1]]%second
        elif(instruction[0]=="eql"):
            variableDict[instruction[1]]= 1 if variableDict[instruction[1]]==second else 0
    print(instruction,variableDict)
    return inputVal
def day24p1(l_instructions,inputVal):
    result=[]
    variableDict=defaultdict(int)
    variableDict["z"]=0
    first=True
    zList=list(str(inputVal))
    #visited=np.zeros(len(l_instructions), dtype=np.bool)
    for i,instruction in enumerate(l_instructions):
        if(instruction[0]=="inp"):
            print(variableDict)
            print("\n")
            if(not first):
                result.append(variableDict["z"])
            first=False
        zList=doInstruction(instruction,variableDict,zList)
    result.append(variableDict["z"])
    print(result)
    return variableDict

def day8p1(l_instructions):
    visited=np.zeros(len(l_instructions), dtype=np.bool)
    acc=0
    i=0
    while i<len(l_instructions):
        if(visited[i]):
            return i,acc,False
        visited[i]=True
        if(l_instructions[i][0]=="acc"):
            acc=max(0,acc+int(l_instructions[i][1]))
        elif(l_instructions[i][0]=="jmp"):
            i=max(0,i+int(l_instructions[i][1]))
            continue
        i+=1
    return i,acc,True


def day8p2(l_instructions):
    checked=np.zeros(len(l_instructions), dtype=np.bool)
    acc=0
    i=0
    end=False
    while not end:
        copy=l_instructions.copy()
        for i in range(len(copy)):
            if(copy[i][0]=="jmp" and not checked[i]):
                copy[i][0]="nop"
                checked[i]=True
                break
            if(copy[i][0]=="nop" and not checked[i]):
                copy[i][0]="jmp"
                checked[i]=True
                break
        test=day8p1(copy)
        if(test[2]):
            end=True
            acc=test[1]
    return acc

l_instructions=generate_list_intructions("input.txt")

day24p1(l_instructions,"99999999999999")
#print(l_instructions)
#99999999999999 11111111111111 13579246899999
# for i in range(99999999999999,11111111111111,-1):
#     if "0" in str(i):
#         continue
#     print(i)
#     testDict=day24p1(l_instructions,i)
#     if(testDict["z"]==0):
#         print(i)
#         print(testDict)
#         break
# print(l_instructions)
#print(day8p1(l_instructions)[1])
#print(day8p2(l_instructions))


```
