```python
def parseinfo(inp):
    whole = inp.split("\n\n")
    rule = lambda x: 0 if x=='.' else 1
    return [*map(rule,whole[0])], np.array([*map(rule,[x for f in whole[1].split('\n') for x in f])]).reshape(100,100)



def solution(inp):
    mapped = np.array(inp[1])
    test=True
    for i in range(3):
        if i== 2: 
            part1 = np.count_nonzero(mapped)
        mapped = np.pad(mapped,1, 'constant', constant_values=(i%2, i%2) )
        step = np.pad(mapped,1, 'constant', constant_values=(i%2, i%2) )
        #print(mapped)
        for x in range(mapped.shape[0]):
            for y in range(mapped.shape[1]):
                #print(x,y)
                gap = step[x:x+3,y:y+3].flatten()
                #print(step[x:x+3,y:y+3],"\n\n")
                #print(gap,"\n", 1 << np.arange(gap.shape[-1]-1,-1,-1)  ) 
                nextrule= gap.dot(1 << np.arange(gap.shape[-1]-1,-1,-1) )
                mapped[x][y] = inp[0][nextrule]
    return part1
                

inp=parseinfo(open(INPUT).read())
solution(inp)
```
