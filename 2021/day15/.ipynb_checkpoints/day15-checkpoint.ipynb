```python
import heapq as heap
import cProfile
EXAMPLE="example.txt"
INPUT="input.txt"

DIRECTION = [(-1,0,),(0,-1),(0,1),(1,0)]

class Edge:
    def __init__(self, v="",x=0):
        self.name = v
        self.n = []
        self.value = x

def initInput(input_,n):
    newInput_=[]
    for element in input_:
        newElement=[]
        for i in range(n):
            temp=[(((x-1)+i)%9)+1 for x in element]
            newElement=newElement + temp
        newInput_.append(newElement)
    copyInput_=newInput_.copy()
    for i in range(1,n):
        generated=[[ (((y-1)+i)%9)+1 for y in x] for x in copyInput_]
        newInput_=newInput_+generated
    return newInput_
        
def getAdjacent(edge,pos,edgeMap,maximums):
    for di in DIRECTION:
        adjacentX,adjacentY=pos[0]+di[0],pos[1]+di[1]
        if(not (0<=adjacentX<maximums[0]) or not (0<=adjacentY<maximums[1])):
            continue
        edge.n.append(edgeMap[adjacentX][adjacentY])

def generateEdgeMap(input_,maximums):
    edgeMap=[]
    for i,element in enumerate(input_):
        line=[]
        for j,e in enumerate(element):
            ed=Edge(v = maximums[1]*i + j, x = e)
            line.append(ed)
        edgeMap.append(line)
    for i,element in enumerate(edgeMap):
        for j,e in enumerate(element):
            getAdjacent(e,[i,j],edgeMap,maximums)
    return edgeMap
             
def dijkstra(dijkstraQ,dist,previous):
    Q,visited=[],[]
    heap.heappush(Q, (0, dijkstraQ[0].name))
    while(Q):
        _,u=heap.heappop(Q)
        if(u==dijkstraQ[-1]):
            break
        for neighbor in dijkstraQ[u].n:
            nValue=neighbor.value
            nDist=dist[dijkstraQ[u].name]+nValue
            if(nDist<dist[neighbor.name]):
                dist[neighbor.name]=nDist
                heap.heappush(Q, (nDist, neighbor.name))
                
def getMinPath(input_):
    maximums=[len(input_),len(input_[0])]
    nbElement=maximums[0]*maximums[1]
    dijkstraMAP=[float("inf")]*nbElement
    dijkstraMAP[0]=0
    previous=[None]*nbElement
    edgeMap=generateEdgeMap(input_,maximums)
    dijkstraQ=[y for x in edgeMap for y in x]
    dijkstra(dijkstraQ,dijkstraMAP,previous)
    return dijkstraMAP[nbElement-1]
    
def day15(fileName):    
    with open(fileName) as f:
        input_=f.read().strip().splitlines()
        input_=[list(map(int,list(x))) for x in input_]

    part2= initInput(input_,5)

    return getMinPath(input_),getMinPath(part2)
cProfile.run('print(day15(INPUT))')
```
