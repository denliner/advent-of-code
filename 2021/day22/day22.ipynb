```python
def generate_list(name):
    l=open(name).read().rstrip().split("\n\n")
    return list(map(int,l[0].split("\n")[1:])),list(map(int,l[1].split("\n")[1:]))

def Round(players,isP2,GameNb,r,nextGameNb):
    #print("-- Round "+str(r)+" (Game "+str(GameNb)+") --")
    #print("Player 1's deck: "+str(players[0]))
    #print("Player 2's deck: "+str(players[1]))
    play=players[0][0],players[1][0]
    players=players[0][1:],players[1][1:]
    #print("Player 1 plays: "+str(play[0]))
    #print("Player 2 plays: "+str(play[1]))
    
    if(isP2 and 
       (len(players[0])>=play[0] and len(players[1])>=play[1])):
        #print("Playing a sub-game to determine the winner...\n")
        _,_,winner,nextGameNb=day22((players[0][:play[0]].copy(),
                                     players[1][:play[1]].copy()) ,isP2,nextGameNb)
        #print("...anyway, back to game "+str(GameNb)+".")
    else:   
        winner=0 if play[0]>play[1] else 1
    #print("Player "+str(winner+1)+" wins round "+str(r)+" of game "+str(GameNb)+"!\n")
    players[winner].append(play[abs(winner)])
    players[winner].append(play[abs(winner-1)])
    return players,winner,nextGameNb


def day22(players,isP2,GameNb):
    #print("=== Game "+str(GameNb)+" ===\n")
    nextGameNb=GameNb+1
    winner,summ,r=0,0,0
    history=[]
    while(players[0] and players[1]):
        r+=1
        players,winner,nextGameNb=Round(players,isP2,GameNb,r,nextGameNb)
        if(isP2):
            if(players in history):
                winner=0
                break
        history.append(players)
    if(GameNb==1):
        for i,element in enumerate(players[winner]):
            #print(str(element)+"* "+ str(len(players[winner])-i))
            summ+=element*(len(players[winner])-i)
    
    return players,summ,winner,nextGameNb
players=generate_list('input.txt')
Part1=day22(players,False,1)
print("Part 1:\t"+str(Part1[1]))
Part2=day22(players,True,1)
print("Part 2:\t"+str(Part2[1]))
```
