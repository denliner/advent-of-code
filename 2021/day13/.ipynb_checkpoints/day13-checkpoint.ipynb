```python
import numpy as np

EXAMPLE="example.txt"
INPUT="input.txt"


def initMap(dotList):
    maxX,maxY=max(dotList[:,0])+1,max(dotList[:,1])+1
    mapu=[]
    for i in range(maxX):
        subMap=[]
        for j in range(maxY):
            subMap.append(".")
        mapu.append(subMap)
    mapu=np.array(mapu)

    for elem in dotList:
        mapu[elem[0]][elem[1]]='#'
    return mapu

def initVariable(fileName):
    with open(fileName) as f:
        input_=f.read().strip().split("\n\n")

    dotList=input_[0].split("\n")
    dotList= [d.split(",") for d in dotList]
    dotList= np.array([[int(x[1]),int(x[0])] for x in dotList ])

    instruction = input_[1].split("\n")
    instruction =[x.split("=") for x in instruction]
    instruction =[[x[0][-1],int(x[1])] for x in instruction]
    mapu=initMap(dotList)
    return instruction,mapu

def flip(instruction,mapu):
    for element in instruction:
        if(element[0]=="x"):
            first=mapu[:,:element[1]]
            second=mapu[:,element[1]+1:]
            for i,elem in enumerate(second):
                for j,e in enumerate(elem):
                    if(e=="#"):
                        first[i][element[1]-(j+1)]='#'
        else:
            first=np.array(mapu[:element[1]])
            second=np.array(mapu[element[1]+1:])
            for i,elem in enumerate(second):
                for j,e in enumerate(elem):
                    if(e=="#"):
                        first[element[1]-(i+1)][j]='#'
        mapu=first
    return mapu

def day13(fileName):
    instruction,mapu=initVariable(fileName)
    silver=flip([instruction[0]],mapu)
    silver=np.count_nonzero(silver == "#") 
    gold=flip(instruction,mapu)
    np.savetxt('gold.txt', gold, fmt='%s')
    return silver

print(day13(INPUT))
```
