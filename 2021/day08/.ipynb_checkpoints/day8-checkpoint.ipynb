```python
import numpy as np

def generate_list_intructions(name):
    return np.genfromtxt(name,dtype=[('U3'),("i8")])
#
def day8p1(l_instructions):
    visited=np.zeros(len(l_instructions), dtype=np.bool)
    acc=0
    i=0
    while i<len(l_instructions):
        if(visited[i]):
            return i,acc,False
        visited[i]=True
        if(l_instructions[i][0]=="acc"):
            acc=max(0,acc+int(l_instructions[i][1]))
        elif(l_instructions[i][0]=="jmp"):
            i=max(0,i+int(l_instructions[i][1]))
            continue
        i+=1
    return i,acc,True


def day8p2(l_instructions):
    checked=np.zeros(len(l_instructions), dtype=np.bool)
    acc=0
    i=0
    end=False
    while not end:
        copy=l_instructions.copy()
        for i in range(len(copy)):
            if(copy[i][0]=="jmp" and not checked[i]):
                copy[i][0]="nop"
                checked[i]=True
                break
            if(copy[i][0]=="nop" and not checked[i]):
                copy[i][0]="jmp"
                checked[i]=True
                break
        test=day8p1(copy)
        if(test[2]):
            end=True
            acc=test[1]
    return acc
    
    
l_instructions=generate_list_intructions("input.txt")
print(l_instructions)
print(day8p1(l_instructions)[1])
print(day8p2(l_instructions))
```

```python
import numpy as np
from collections import defaultdict

def generate_list_intructions(name):
    return np.genfromtxt(name,dtype=['U3',"i8"])

def day8p1(l_instructions):
    visited=set()
    acc=0
    i=0
    nopjmp=set()
    lenght=len(l_instructions)
    while i not in visited and i<lenght :
        visited.add(i)
        instruction=l_instructions[i]
        # to get the index of all nop and jmp instruction encountered during the loop
        if(instruction[0]=="nop" or instruction[0]=="jmp"):
            nopjmp.add(i)        
        if(instruction[0]=="acc"):
            acc=acc+int(instruction[1])
            i+=1
        elif(instruction[0]=="jmp"):
            i=i+int(instruction[1])
        else:
            i+=1
    if(i==lenght):
        return i,acc,True,nopjmp
    else:
        return i,acc,False,nopjmp


def getAllPath(l_instructions):
    lenght=len(l_instructions)
    path=defaultdict(set)
    for i in range(lenght):
        s=i
        next_i=i+1
        if(l_instructions[i][0]=="jmp"):
            next_i=i+l_instructions[i][1]
        path[next_i].add(s)
    return path

def targetNeighborhood(l_instructions,path):
    target_N=set()
    current_inst={len(l_instructions)}
    while current_inst:
        copy=current_inst.pop()
        target_N.add(copy)
        current_inst = current_inst.union(path[copy] - target_N)
    return target_N
            

    

def day8p2(l_instructions):
    length=day8p1(l_instructions)[3] #indexes of all nop/jmp instruction encountered
    target_N=targetNeighborhood(l_instructions,getAllPath(l_instructions))
    for i in length :
        if(l_instructions[i][0]=="jmp" and (i+1) in target_N):
            l_instructions[i][0]="nop"
            test=day8p1(l_instructions)
            l_instructions[i][0]="jmp"
            if(test[2]):
                return test[1]
        elif(l_instructions[i][0]=="nop"  and (i+l_instructions[i][1]) in target_N):
            l_instructions[i][0]="jmp"
            test=day8p1(l_instructions)
            l_instructions[i][0]="nop"
            if(test[2]):
                return test[1]
    return 0
    
l_instructions=generate_list_intructions("jfnwh2.txt")
print(day8p1(l_instructions)[1])
print(day8p2(l_instructions))
```

```python
def proc2b(data):
    # Give op, arg pairs
    code = generate_list_intructions(data)

    # Graph of all instructions that lead to a particular instruction.
    rev_jmp_tbl = defaultdict(set)

    for pc, (op, arg) in enumerate(code):
        src = pc
        dst = pc + 1
        if op == "jmp":
            dst = pc + arg
        rev_jmp_tbl[dst].add(src)

    # Traverse the graph to find all nodes leading to the end. The patched code
    # must result in jumping into this set.
    targets = set()
    edges = {len(code)}
    while edges:
        node = edges.pop()
        targets.add(node)
        edges |= rev_jmp_tbl[node] - targets

    # Run the machine, but check for the first jmp or nop which when patched
    # will put the pc into the target set.
    pc = 0
    acc = 0
    patched = False
    while pc != len(code):
        op, arg = code[pc]
        if not patched:
            if op == "nop" and (pc + arg) in targets:
                op = "jmp"
                patched = True
            elif op == "jmp" and (pc + 1) in targets:
                op = "nop"
                patched = True
        if op == "acc":
            acc += arg
            pc += 1
        elif op == "nop":
            pc += 1
        elif op == "jmp":
            pc += arg
    print(acc)
cProfile.run("proc2b('jfnwh2.txt')")
```
