```python
from collections import defaultdict
from cProfile import run

EXAMPLE="example.txt"
sEXAMPLE="smallExample.txt"
INPUT="input.txt"
BIGBOY="8-100000.in"
def diff(x,y):
    setX=set(x)
    return len([item for item in y if item not in setX])


def getValues(patterns):
    val={}
    for v in patterns:
        l=len(v)
        if(l==2):
            val[1]=v
        elif(l==4):
            val[4]=v
        elif(l==3):
            val[7]=v
        elif(l==7):
            val[8]=v
            
    for v in patterns:
        l=len(v)
        if(l==5):
            if(diff(val[1],v)==3):
                val[3]=v
            elif(diff(val[4],v)==2):
                val[5]=v
            else:
                val[2]=v
        elif(len(v)==6):
            if(diff(val[1],v)==5):
                val[6]=v
            else: 
                if(diff(val[4],v)==3):
                    val[0]=v
                else:
                    val[9]=v
    return val



def day08(input_):
    count= defaultdict(int)
    sFinal=[]
    for l in input_:
        s=''
        values=getValues(l[0])
        translation= { "".join(sorted(values[n])):n for n in values}
        for elem in l[1]:
            num=translation["".join(sorted(elem))]
            s+=str(num)
            count[num]+=1
        sFinal.append(int(s))
    return sum([count[x] for x in [1,4,7,8]]),sum(sFinal)


with open(BIGBOY) as f:
    input_=f.read().splitlines()

input_=[x.split(" | ") for x in input_]
input_=[[x.split(" ") for x in y] for y in input_]
run("print(day08(input_))")

#         8800042 function calls in 2.685 seconds
```
