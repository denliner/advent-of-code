```python
import numpy as np

def generate_list(name):
    return np.genfromtxt(name,dtype="i8")

def generate_preamble(data,n):
    return data[0:n]

def day1p1(liste,z,length):
    i=0
    j=length-1
    while(i<j):
        if((liste[i]+liste[j])==z):
            return liste[i]*liste[j]
        elif((liste[i]+liste[j])<z):
            i+=1
        else:
            j-=1
    return None

def getContignousSet(data,target):
    length=len(data)
    for i in range(length):
        contiguous=np.array([],dtype="i8")
        num=data[i]
        if(num<target):
            y=target-num
            contiguous=np.append(contiguous,num)
            for j in range(i+1,length):
                num2=data[j]
                if(num2<y):
                    contiguous=np.append(contiguous,num2)
                    y=y-num2
                elif(num2==y):
                    return np.append(contiguous,num2)
                else:
                    break
    return None
                    
        
def day9p1(data,n):
    preamble=generate_preamble(data,n)
    length=len(data)
    for i in range(n,length):
        if day1p1(np.sort(preamble),data[i],n) is None:
            return data[i]
        preamble=np.append(np.delete(preamble,0),data[i])
    return None

def day9p2(data,target):
    contiguous=getContignousSet(data,target)
    print(contiguous)
    if(contiguous is not None):
        return min(contiguous)+max(contiguous)


data=generate_list("input.txt")
print(day9p2(data,day9p1(data,25)))
#13549369
```

```python
def generate_list(name):
    return list(map(int,open(name).read().splitlines()))

def generate_preamble(data,n):
    return data[0:n].copy()

def day1p1(liste,z,length):
    i=0
    j=length-1
    while(i<j):
        if((liste[i]+liste[j])==z):
            return True
        elif((liste[i]+liste[j])<z):
            i+=1
        else:
            j-=1
    return None

def day9p1(data,n):
    preamble=generate_preamble(data,n)
    length=len(data)
    for i in range(n,length):
        if day1p1(sorted(preamble),data[i],n) is None:
            print("Part 1:\n"+str(data[i])+"\n\n")
            return data[i]
        preamble.pop(0)
        preamble.append(data[i])
    return None

def getContignousSet(data,target):
    contiguous=list()
    y=0
    for num in data:
        y+=int(num)
        contiguous.append(num)
        while(y>target):
            y-= int(contiguous[0])
            contiguous=contiguous[1:]
    
        if(y == target and len(contiguous)>=2):
            return contiguous
        
    return None
                    
def day9p2(data,target):
    contiguous=getContignousSet(data,target)
    if(contiguous is not None):
        contiguous = [int(x) for x in contiguous]
        return min(contiguous)+max(contiguous)

data=generate_list("bigboy9c.txt")
target=day9p1(data,25)
print("Part 2=\n"+str(day9p2(data,target))+"\n\n")

#13549369
```
