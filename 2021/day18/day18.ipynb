```python
EXAMPLE="example.txt"
INPUT="input.txt"
import math
import ast
import cProfile


class Element:
    def __init__(self, v="",x=0,isLeft=True, p= None):
        self.name = v
        self.value = x
        self.isLeft = isLeft
        self.parent=p
class Pair:
    def __init__(self, v="", p= None,isLeft=True):
        self.name = v
        self.first = None
        self.second = None
        self.parent=p
        self.isLeft= isLeft


def doTree(arrr,parent=None):
    parent=Pair(str(arrr),parent)
    #print(doList2(arrr))
    parent.first=doTree(arrr[0],parent) if isinstance(arrr[0], list) else Element(str(arrr[0]),x=arrr[0], p=parent) 
    parent.first.isLeft=True
    parent.second=doTree(arrr[1],parent) if isinstance(arrr[1], list) else Element(str(arrr[1]),x=arrr[1], p=parent) 
    parent.second.isLeft=False
    return parent


def printList(arrr):
    first=arrr.first.value if isinstance(arrr.first, Element) else printList(arrr.first)
    second=arrr.second.value if isinstance(arrr.second, Element) else printList(arrr.second)
    newArrr=[first,second]
    return newArrr

def getElemToExplode(arrr,n):
    if(n==4):
        if isinstance(arrr.first,Pair):
            return arrr.first
        elif isinstance(arrr.second,Pair):
            return arrr.second
    else:
        if isinstance(arrr.first,Pair):
            found=getElemToExplode(arrr.first,n+1)
            if(found):
                return found
        if isinstance(arrr.second,Pair):
            found=getElemToExplode(arrr.second,n+1)
            if(found):
                return found
    return None
            

def getElemToSplit(arrr):
    listElement=[]
    if isinstance(arrr.first,Pair):
        found=getElemToSplit(arrr.first)
        if(found):
            #print(found.name)
            return found
    else:
        if(arrr.first.value>=10):
            return arrr.first
    
    if(isinstance(arrr.second,Pair)):
        found=getElemToSplit(arrr.second)
        if(found):
            #print(found.name)
            return found
    else:
        if(arrr.second.value>=10):
            return arrr.second
    #print("arrr.name",arrr.name)
    return None


def getMagnitude(combine):
    first=combine[0] if isinstance(combine[0], int) else getMagnitude(combine[0])
    second=combine[1] if isinstance(combine[1], int) else getMagnitude(combine[1])
    newArrr=first*3+second*2
    return newArrr


def decentToFirstLR(elem,isLeft):
    if(isinstance(elem,Pair)):
        nextt=elem.first if isLeft else elem.second
        if(isinstance(nextt,Element)):
            return nextt
        else:
            return decentToFirstLR(nextt,isLeft)
    else:
        return elem


def mountToFirstLR(elem,isLeft):
    if(isinstance(elem,Pair)):
        nextt=elem.parent.first if isLeft else elem.parent.second
        if(nextt==elem):
            if(elem.parent.parent is None):
                return None
            else:
                return mountToFirstLR(elem.parent,isLeft)
        else:
            return decentToFirstLR(nextt,not isLeft)


        

def explodeList(listPair,toExplode):
    #print(printList(listPair))
    if(toExplode):
        nextLeft=mountToFirstLR(toExplode,True)
        if(nextLeft is not None):
            nextLeft.value=nextLeft.value+toExplode.first.value
            nextLeft.name=str(nextLeft.value)
        nextRight=mountToFirstLR(toExplode,False)
        if(nextRight is not None):
            nextRight.value=nextRight.value+toExplode.second.value      
            nextRight.name=str(nextRight.value)      
        if toExplode.isLeft:
            toExplode.parent.first=Element("0",x=0,isLeft=True, p= toExplode.parent)
        else:
            toExplode.parent.second=Element("0",x=0,isLeft=False, p= toExplode.parent)
        #toExplode.value
        #print(printList(listPair))
        return listPair
    else:
        return None
    
    
    
    
def splitList(listPair,toSplit):
    #print(printList(listPair))
    if(toSplit):    
        x,y= math.floor(toSplit.value/2), math.ceil(toSplit.value/2)
        newPair= Pair(v= str([x,y]),p=toSplit.parent,isLeft=toSplit.isLeft)
        newPair.first=Element(str(x),x=x,isLeft=True, p= newPair)
        newPair.second=Element(str(y),x=y,isLeft=False, p= newPair)    
        if toSplit.isLeft:
            toSplit.parent.first=newPair
        else:
            toSplit.parent.second=newPair
        #print(printList(listPair))
        return listPair
    else:
        return None
def snailAddition(list1,list2):
    combine=[]
    combine.append(list1)
    combine.append(list2)
    return combine
    
def combineAll(input_):
    combine=input_[0]
    for i in range(1,len(input_)):
        combine=snailAddition(combine,input_[i])
        #print(combine)
        while(1):
            listPair=doTree(combine)
            toExplode=getElemToExplode(listPair,1)
            toSplit=getElemToSplit(listPair)
            if(toExplode):
                listPair=explodeList(listPair,toExplode)
            elif(toSplit):
                listPair=splitList(listPair,toSplit)
            else:
                break
            combine=printList(listPair)
        #print("\n"*4)
    return combine

def getSum(input_):
    combine=combineAll(input_)
    return getMagnitude(combine)

def day18(fileName):
    with open(fileName) as f:
        input_=f.read().strip().splitlines()
    input_=[ast.literal_eval(x) for x in input_]
    silver=getSum(input_)
    gold=0
    allCouple=[]
    for x in input_:
        for y in input_:
            if(not x==y):
                allCouple.append([x,y])
    for couple in allCouple:
        summ=getSum(couple)
        if(summ>gold):
            gold=summ
    return silver,gold
                
#explodeList(test3)
#day18(EXAMPLE)
#snailAddition([1,2],[[3,4],5])
#with open(EXAMPLE) as f:
#    input_=f.read().strip().splitlines()
#    input_=[eval(x) for x in input_]
#combineAll(input_)
#getMagnitude()
cProfile.run("print(day18(INPUT))")
#mountToFirstLeft(p.first.first.first.first),p.name
```
