```python
import re
import numpy as np
import cProfile
EXAMPLE="example.txt"
INPUT="input.txt"

with open(INPUT) as f:
    input_=list(map(int, list(re.findall('\-?\d+', f.read() ))))


def distanceCostP1(crab,newPosition):
    return abs(crab-newPosition)

def distanceCostP2(crab,newPosition):
    n=distanceCostP1(crab,newPosition)
    return (pow(n,2)+n)//2

def getFuelCost(crabPos,pos,maximum,distanceCost):
    fuel=0
    for elem in crabPos:
        fuel+=distanceCost(elem,pos)
        if(fuel>maximum):
            return float('+inf')
    return fuel

def getMinMax(input_,distanceCost):
    minFuel=float('+inf')
    low=min(input_)
    up=max(input_)+1
    for elem in range(low,up):
        cost=getFuelCost(input_,elem,minFuel,distanceCost)
        if(cost<minFuel):
            minFuel=cost
    return minFuel

cProfile.run("print(getMinMax(input_,distanceCostP1),getMinMax(input_,distanceCostP2))")
```

```python
from statistics import mean,median
import re
import cProfile
EXAMPLE="example.txt"
INPUT="input.txt"
BIGBOY="7-1000000-2.in"
with open(INPUT) as f:
    input_=list(map(int, list(re.findall('\-?\d+', f.read() ))))

def distanceCostP1(crab,newPosition):
    return abs(crab-newPosition)

def distanceCostP2(crab,newPosition):
    n=distanceCostP1(crab,newPosition)
    return (pow(n,2)+n)//2

def getFuelCost(crabPos,pos,distanceCost):
    fuel=0
    for elem in crabPos:
        fuel+=distanceCost(elem,pos)
    return fuel
def getPosP1(crabPos):
    return round(median(crabPos))

def getPosP2(crabPos):
    return round(mean(crabPos))

cProfile.run("print(getFuelCost(input_,getPosP1(input_),distanceCostP1),getFuelCost(input_,getPosP2(input_),distanceCostP2))")
```
