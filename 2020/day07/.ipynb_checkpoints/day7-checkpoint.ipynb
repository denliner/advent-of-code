```python
def countGold(bag,lbags):
    nbGold=0
    if not bag["contain"]:
        return 0
    if "shiny gold bags" in bag["contain"]:
        return 1
    for keys in bag["contain"]:
        if(lbags[keys]["shiny gold"]==-1):
            nbGold+=countGold(lbags[keys],lbags)
            if(nbGold>=1):
                return nbGold
            

    bag["shiny gold"]=nbGold
    return nbGold

def countBag(bag,lbags):
    nbBag=0
    if isinstance(bag["contain"],int):
        return bag["contain"]
    for keys in bag["contain"]:
        nbBag+=(1+countBag(lbags[keys],lbags))*bag["contain"][keys]
    bag["contain"]=nbBag
    return nbBag


def day7p1(lbags):
    n=0
    for key in lbags:
        if  lbags[key]["shiny gold"] == -1:
            countGold(lbags[key],lbags)
        if(not lbags[key]["shiny gold"]==0):
            n+=1
    return n

def generateDict(inp):
    with open(inp) as f:
        lbags = f.read().strip().split("\n")
    lbags=[sub.split(" contain ") for sub in lbags]
    for i in range(len(lbags)):
        lbags[i][1]=lbags[i][1].split(", ")
        dictionary={"contain": {},"shiny gold": -1}
        for j in range(len(lbags[i][1])):
            entry=lbags[i][1][j].split()
            if entry[0].isalpha():
                dictionary["shiny gold"]=0
                break
            dictionary["contain"][(" ".join(entry[1:])).replace(".","")+"s"* (1 if int(entry[0]) == 1 else 0)]=int(entry[0])
        lbags[i][1]=dictionary
    bagdict={}
    for bag in lbags:
        bagdict[bag[0]]=bag[1]
    
    return bagdict
    

lbags=generateDict("2lj4p8.txt")
print("Part 1:")
print(day7p1(lbags))
print("Part 2:")
print(countBag(lbags["shiny gold bags"],lbags))

#print(json.dumps(lbags, indent=2))
#lbags= [sub.split(", ") for subl in lbags for sub in subl]
#lbags=np.array([np.array(sub.split(" ")) for subll in lbags for subl in subll for sub in subl])
#lbags[0][1]['mirrored fuchsia bags']
#lbags["faded yellow bags"]["shiny gold"]
#lbags 257
```
