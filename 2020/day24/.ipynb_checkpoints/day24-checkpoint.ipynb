```python
directionDict={"w": (-2,0),"e":(2,0),"nw":(-1,1),
              "sw":(-1,-1),"ne":(1,1),"se":(1,-1)}

def generate_hexlist(name):
    return open(name).read().rstrip().split("\n")
hexlist=generate_hexlist("input.txt")

def day24p1(hexlist):
    flipped=0
    blacks={}
    for tile in hexlist:
        condition=False
        position=(0,0)
        for i,e in enumerate(tile):
            if(condition):
                condition=False
                continue
            direction=e
            if(e in ['s','n']):
                condition=True
                direction=e+tile[i+1]
            position=tuple(map(sum, zip(position, directionDict[direction])))
        if(position not in blacks):
            flipped+=1
            blacks[position]=True
        else:
            flipped-=1
            blacks.pop(position, None)
    return blacks,flipped

blacks,flipped=day24p1(hexlist)
print(flipped)

def countBlackNeighbours(blacks,currentTile):
    count=0
    whiteboi={}
    for di in directionDict.values():
        #print(currentTile)
        #print(di)
        #print("\n")
        n=tuple(map(sum, zip(currentTile, tuple(di))))
        if(n in blacks):
            count+=1
        else:
            #print(n)
            whiteboi[n]=True
    return count,whiteboi

def day24p2(blacks,flipped,limit):
    i=0
    while i<limit:
        karaBoga=blacks.copy()
        whiteboi={}
        for element in blacks:
            count, white=countBlackNeighbours(blacks,element)
            whiteboi|=white
            if (count==0 or count>2):
                karaBoga.pop(element,None)
                flipped-=1
        #print(blacks-whiteboi)
        for element in whiteboi:
            count,_=countBlackNeighbours(blacks,element)
            if(count==2):
                karaBoga[element]=True
                flipped+=1
        blacks=karaBoga
        #print(i+1,flipped)
        i+=1
    return flipped

day24p2(blacks,flipped,100)
```
