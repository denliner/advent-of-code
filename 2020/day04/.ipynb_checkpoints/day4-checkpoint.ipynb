```python
import re
FIELDS=["byr","iyr","eyr","hgt","hcl","ecl","pid"]
#[1-9a-f]*
def day4p1(lpassword):
    n=0
    for passports in lpassword:
        n+=validate_passports(passports)
    return n
    
    
def validate_passports(passports):
    if(len(passports)<7):
        return 0
    for field in FIELDS:
        if (not field in passports.keys()):
            return 0
    return 1

def validate_years(value,up,down):
    try:
        return (down<=int(value)<=up)
    except ValueError:
        return False
def validate_height(value):
    try:
        if "cm" in value:
            return (150<=int(value[:-2])<=193)
        elif "in" in value:
            return (59<=int(value[:-2])<=76)
        else:
            return False
    except ValueError:
        return False
def validate_hair(color):
    return bool(re.compile("#[0-9a-f]{6}$").match(color))

def validate_eyes(eyecolor):
    if(not len(eyecolor)==3):
        return False
    return any(n in eyecolor for n in ["amb","blu","brn","gry","grn","hzl","oth"])

def validate_pid(pid):
    return bool(re.compile("[0-9]{9}$").match(pid))

def validate_passportsP2(passports):
    if( not (7<=int(len(passports))<=8)):
        return 0
    for field in FIELDS:
        if (not field in passports.keys()):
            return 0
    return int(validate_years(passports["byr"],2002,1920)
               and validate_years(passports["iyr"],2020,2010)
               and validate_years(passports["eyr"],2030,2020)
               and validate_height(passports["hgt"])
               and validate_hair(passports["hcl"])
               and validate_eyes(passports["ecl"])
               and validate_pid(passports["pid"]))


def day4p2(lpassword):
    n=0
    for passports in lpassword:
        n+=validate_passportsP2(passports)
    return n

with open('input.txt') as f:
    lpassword = f.read().replace("\n\n", "\t").replace(" ", ":").replace("\n", ":").split("\t")

lpassword = [sub.split(":") for sub in lpassword] 
lpassword = [dict(zip(sub[::2], sub[1::2])) for sub in lpassword]
#print(lpassword)
#print(day4p1(lpassword))
day4p2(lpassword)
#lpassword = [sub.split(":") for subl in lpassword for sub in subl] 

#lpassword = [dict([sub]) for sub in lpassword] 

#lpassword = [[sub.split(":")][0] for subl in lpassword for sub in subl] 
```
