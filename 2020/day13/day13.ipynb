```python
def powmod(b, e, m, c=1):
    b %= m
    e %= m
    while e > 0:
        if e % 2 == 1:
            c *= b
            c %= m
        e //= 2
        b *= b
        b %= m
    return c % m
powmod(143,7-2,7)
```
