```python
import math
def getSeat(seats):
    row=0
    upperRow=(2**7)-1
    lowerRow=0
    col=0
    upperCol=(2**3)-1
    lowerCol=0
    for character in seats:
        if character=='F':
            row=upperRow=upperRow - math.ceil((upperRow-lowerRow)/2)
        elif character=='B':
            row=lowerRow=lowerRow + math.ceil((upperRow-lowerRow)/2)
        elif character=='L':
            col=upperCol=upperCol - math.ceil((upperCol-lowerCol)/2)
        elif character=='R':
            col=lowerCol=lowerCol + math.ceil((upperCol-lowerCol)/2)
        else:
            print("Kill yourself")
    return (row*8+col)

def getSeatLazy(seats):
    return  int(seats.replace("F", "0").replace("B", "1").replace("L", "0").replace("R", "1"),2)
 

def getMaxSeatID(lseats):
    maxID=0
    seatsID=0
    for seats in lseats:
        seatsID=getSeat(seats)
        if maxID<seatsID:
            maxID=seatsID
    return maxID

    
def getMaxSeatIDP2(lseats):
    maxID=0
    lSeatsID=[]
    seatsID=0
    for seats in lseats:
        lSeatsID.append(getSeatLazy(seats))
    
    missing= [mis for mis in range(min(lSeatsID),max(lSeatsID) + 1) if mis not in lSeatsID]
    print(missing)
    return missing    

def getMaxSeatIDP2N(lseats):
    maxID=0
    lSeatsID= np.zeros(len(lseats),dtype="int")
    seatsID=0
    #print(range(len(lseats)))
    for i in range(len(lseats)):
        lSeatsID[i]=getSeatLazy(lseats[i])
    
    #print(lSeatsID)
    #print(np.amin(lSeatsID))
    #print(np.amax(lSeatsID))
    #print(range(len(lSeatsID)))
    #print(np.amax(lSeatsID))
    #lSeatsID=np.unique(np.sort(lSeatsID, axis=None))
    #print(lSeatsID)
    #print(np.amin(lSeatsID))
    #print(np.amax(lSeatsID))
    missing=np.setdiff1d(np.arange(np.min(lSeatsID),np.max(lSeatsID) + 1),lSeatsID)
    #missing= [mis for mis in np.arange(np.min(lSeatsID),np.max(lSeatsID) + 1) if mis not in lSeatsID]

    #for mis in np.array(range(np.min(lSeatsID),np.max(lSeatsID) + 1)):
    #    if mis not in lSeatsID:
    #        np.append(missing,mis)
    print(missing)
    #print(len(missing))
    #print(len(range(np.min(lSeatsID),np.max(lSeatsID) + 1)))
    return missing

with open('input.txt') as f:
    lseats = f.read().splitlines()


print(lseats[0])    
len(lseats)
getMaxSeatIDP2N(lseats)
int("10011010010000000111",2)
```
