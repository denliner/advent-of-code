```python
import math,re,heapq,operator,sys
from copy import deepcopy
reg="-?[0-9]+"
from collections import defaultdict,deque


INPUT="input.txt"
LOWCOUNT=0
HIGHCOUNT=0



with open(INPUT) as f:
    inp=f.read().strip().splitlines()
    inp=[x.split(" -> ") for x in inp]


Gold=0

def modWork(dictMod,limit,isP2=False,output=""):
    if(isP2):
        goal=dictMod[output]["inp"][0]
        goalPred=dictMod[goal]["inp"]
        goalCount=len(goalPred)
        goalList=[]
    global LOWCOUNT,HIGHCOUNT
    for i in range(limit):
        # print("\n")
        LOWCOUNT+=1
        d=deque()
        d.append(("broadcaster",False))
        while d:
          #  print(d)
            modKey,inpulse=d.popleft()
            if(isP2 and modKey==goal and inpulse==True):
                print(i,modKey,inpulse,[dictMod[x]["pulse"] for x in goalPred])
                goalList.append(i+1)
                if(len(goalList)==goalCount):
                    global gold
                    print(goalList)
                    gold= math.prod(goalList)
                    return
            if(not modKey in dictMod):
                continue
            match dictMod[modKey]["type"]:
                case "broadcaster":
                    dictMod[modKey]["pulse"]=False
                    # for elem in dictMod[modKey]["output"]:
                    #     dictMod[elem]["pulse"]=False
                    #     print("boardcaster",dictMod[elem]["pulse"],elem)
                    for elem in dictMod[modKey]["out"]:
                        d.append((elem,False))
                        LOWCOUNT+=1
                        # print("boardcaster",False,elem,inpulse)
                        # modWork(dictMod,elem,dictMod[elem]["pulse"])
                case "%":
                    if(not inpulse):
                        nPulse=dictMod[modKey]["pulse"] 
                        nPulse=nPulse if inpulse else not nPulse
                        dictMod[modKey]["pulse"]=nPulse
                        for elem in dictMod[modKey]["out"]:
                            if(nPulse):
                                HIGHCOUNT+=1
                            else:
                                LOWCOUNT+=1
                            # print("%",modKey,nPulse,elem,inpulse)
                            d.append((elem,nPulse))
                case "&":
                    nPulse=False
                    for elem in dictMod[modKey]["inp"]:
                        if (dictMod[elem]["pulse"]==False or inpulse==False):
                            nPulse=True
                            break
                    dictMod[modKey]["pulse"]=nPulse
                    for elem in dictMod[modKey]["out"]:
                        if(nPulse):
                            HIGHCOUNT+=1
                        else:
                            LOWCOUNT+=1
                        # print("&",modKey,nPulse,elem,inpulse,dictMod[modKey]["inp"])
                        d.append((elem,nPulse))
                        # modWork(dictMod,elem,nPulse)
                case "output":
                    continue



dictModule=defaultdict(dict)
conjInput=defaultdict(list)
for elem in inp:
    type=""
    name=""
    pulse=False
    if(elem[0]=="broadcaster"):
        type="broadcaster"
        name=elem[0]
    elif(elem[0][0]=="%"):
        type="%"
        name=elem[0][1:]
    elif(elem[0][0]=="&"):
        type="&"
        name=elem[0][1:]
        pulse=True
    else:
        type="output"
        name=elem[0]
        pulse=False        
    output=elem[1].split(", ")
    for o in output:
        conjInput[o].append(elem[0])
    dictModule[name]["out"]=output
    dictModule[name]["type"]=type
    dictModule[name]["pulse"]=pulse
    dictModule[name]["inp"]=[]
#print(dictModule)
dictModule["rx"]["out"]=[]
dictModule["rx"]["type"]="output"
dictModule["rx"]["pulse"]=False
dictModule["rx"]["inp"]=[]
dictModule
for key in dictModule:
    for out in dictModule[key]["out"]:
        if(out in dictModule):
            # if(dictModule[out]["type"]=="&"):
           # print(out,dictModule[out])
            dictModule[out]["inp"].append(key)


modWork(deepcopy(dictModule),1000)
silver=LOWCOUNT*HIGHCOUNT
print(silver)
modWork(deepcopy(dictModule),100000,True,"rx")

# #dictModule
#dictModule
#reverse(dictModule,'rx')
#dictModule
silver,gold
```
