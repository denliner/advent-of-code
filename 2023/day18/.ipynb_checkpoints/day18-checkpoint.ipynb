```python
import math,re,heapq,operator,sys
import numpy as np
from functools import cache,lru_cache
from copy import deepcopy
sys.setrecursionlimit(100000000)

from collections import defaultdict
INPUT="input.txt"
EXGrid="exampleGrid.txt"
dirr= {"north": (-1,0),"south": (1,0),
       "east": (0,1),"west": (0,-1), }

conversion= {"U" :"north", "D":"south", "R":"east", "L":"west"}
#inst={}
#gridD={}
with open(INPUT) as f:
    inp=f.read().strip().splitlines()
    inp=[x.replace("(","").replace(")","").split(" ") for x in inp]

for i in range(len(inp)):
    inp[i][0]=conversion[inp[i][0]]
    inp[i][1]=int(inp[i][1])

def digV2(instructions):
    index=(0,0)
    rect=[]
    digLen=0
    for elem in instructions:
        rect.append(index)
        direction=dirr[elem[0]]
        count=elem[1]
        vect= tuple(x * count for x in direction)
        index=tuple(map(sum, zip(index, vect)))
        digLen+=count
        # for i in range(count):
        #     digLen+=1
        #     index=tuple(map(sum, zip(index, direction)))
        #    # gridD[index]=True
        # if(not index in rect):
        #     rect.append(index)
        # maxX=index[0] if index[0]>maxX else maxX
        # maxY=index[1] if index[1]>maxY else maxY

        # minX=index[0] if index[0]<minX else minX
        # minY=index[1] if index[1]<minY else minY
    return rect,digLen

def shoelace(p,l):
    return 0.5 * abs(sum(x0*y1 - x1*y0
                         for ((x0, y0), (x1, y1)) in segments(p))) + l//2 + 1

def segments(p):
    return zip(p, p[1:] + [p[0]])


def dig(instructions):
    gridD={}
    index=(0,0)
    maxX,maxY=0,0
    minX,minY=0,0
    digLen=0
    rect=[(0,0)]
    for elem in instructions:
        direction=dirr[elem[0]]
        count=elem[1]
        val=elem[2]
        for i in range(count):
            digLen+=1
            index=tuple(map(sum, zip(index, direction)))
           # gridD[index]=True
        if(not index in rect):
            rect.append(index)
        maxX=index[0] if index[0]>maxX else maxX
        maxY=index[1] if index[1]>maxY else maxY

        minX=index[0] if index[0]<minX else minX
        minY=index[1] if index[1]<minY else minY
    return gridD,minX,minY,maxX,maxY,rect,digLen

# with open(EXGrid) as f:
#     grid=f.read().strip().splitlines()

#nGrid=deepcopy(grid)
#print("\n".join(grid))

#nGrid=[list(x) for x in nGrid]

def DFS(index,setInter,maxX,maxY,visited,minX,minY,inpC,oldInter):
    out=False
    for elem in dirr.values():
        neigh=tuple(map(sum, zip(index, elem)))
        x,y=neigh
        if(neigh in oldInter):
            return True
        if(minX <= x < maxX and minY <= y < maxY):
            setInter.add(index)
            #print("fuck")
            #print(neigh)
        #    print(x,y,inpC[x][y],inpC[index[0]][index[1]])
            if(inpC[x][y]=="#"):
               # print(neigh)
                continue
            # else :
            #     if(neigh in graph):
            #         setInter=set()
            #         return False
            if((x,y) in visited ):
                continue
            visited.add((x,y))

            t=DFS(neigh,setInter,maxX,maxY,visited,minX,minY,inpC,oldInter)
            if(t):
                out=True
        else:
            out=True
       #     print("fuck2")
    return out

def doPart(inpp):
    dictG,minX,minY,maxX,maxY,rect,digLen=dig(inpp)
    nDictG={}
    #print(rect)
    if(minX<0 or minY<0):
        maxX-=minX
        maxY-=minY
        for elem in dictG:
            newIndex=(elem[0]-minX,elem[1]-minY)
            nDictG[newIndex]=dictG[elem]
    else:
        nDictG=deepcopy(dictG)
    print(maxX,maxY)
    grid = [["." for _ in range(maxY+1)] for _ in range(maxX+1)]
    print(len(grid),len(grid[0]))
    for elem in nDictG:
        #print(elem)
        grid[elem[0]][elem[1]]="#"
    
    nGrid=deepcopy(grid)
    
    
    
    finalInter=set()
    visited=set()
    
    for i in range(len(nGrid)):
        for j in range(len(nGrid[0])):
            if(nGrid[i][j]=="."):
                if((i,j) in finalInter ):
                    continue
                #print((i,j))
                setInter=set()
                Cubes=[]
                result=DFS((i,j),setInter,
                           len(nGrid),len(nGrid[0]),
                           visited,0,0,nGrid,finalInter)
               # print(result)
                if(result):
                    finalInter=finalInter | setInter
    
    
    for elem in finalInter:
        nGrid[elem[0]][elem[1]]=" "
    
    for i in range(len(nGrid)):
        for j in range(len(nGrid[0])):
            if(nGrid[i][j]=="."):
                nGrid[i][j]="#"
    
    silverCount=0
    for line in nGrid:
        for elem in line:
            if(elem=="#"):
                silverCount+=1
    # nGrid=["".join(x) for x in nGrid]
    # print("\n".join(nGrid))
    return silverCount


convertP2={"3" :"north", "1":"south", "0":"east", "2":"west"}

def convertTo(oldInp):
    hexx=oldInp[2]
    di=convertP2[hexx[-1]]
    h=int(hexx[1:-1],16)
    return [di,h,hexx]

newInp=[]

for elem in inp:
    newInp.append(convertTo(elem))


rect,digLen=digV2(inp)
silver=int(shoelace(rect,digLen))

rect,digLen=digV2(newInp)

gold=int(shoelace(rect,digLen))
silver,gold
```
