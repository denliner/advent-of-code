```python
import re,cProfile
from collections import defaultdict
from functools import cache
reg="-?[0-9]+"
INPUT="input.txt"
#0 upward 
#1 downward
#2 leftward
#3 rightward
mirrorDict= { 
    "/": {3: 0,
          2: 1,
          0:3,
          1:2 },
    "\\": {3: 1,
          2: 0,
          0:2,
          1:3 }
}
dirr= {3: (0,1),
          2: (0,-1),
          0: (-1,0),
          1:(1,0) }


with open(INPUT) as f:
    inp=f.read().strip().splitlines()
    #inp= [list(x) for x in inp]
inp

mirrorL={}
splitterL={}
for i,line in enumerate(inp):
    for j,elem in enumerate(line):
        if(elem in ["/","\\"]):
            mirrorL[(i,j)]=elem
        elif(elem in ["-","|"]):
            splitterL[(i,j)]=elem
def mirror(lightRay,type):
    newRay=(lightRay[0],mirrorDict[type][lightRay[1]])
    return newRay

def splitter(lightRay,type):
    if(type=="|"):
        if(lightRay[1] in [0,1] ):
            return [lightRay]
        else:
            newRay1= (lightRay[0],0)
            newRay2= (lightRay[0],1)
            return [newRay1,newRay2]
    if(type=="-"):
        if(lightRay[1] in [2,3] ):
            return [lightRay]
        else:
            newRay1= (lightRay[0],3)
            newRay2= (lightRay[0],2)     
            return [newRay1,newRay2]
    return None

def part1(energized,rayList,dirr,mapp):
    maxX,maxY= len(mapp),len(mapp[0])
    count=0
    start=True
    seen=set()
    maxCount=(maxX+maxY)
    previousEner=len(energized.keys())
    while(len(rayList)>0):
        newList=[]
        for i,elem in enumerate(rayList):
            if(elem in seen):
                continue
            if(not start):    
                seen.add(elem)
            if(start):
                newRay=(elem[0],elem[1])
                start=False
            else:
                newRay=(tuple(map(sum, zip(elem[0], dirr[elem[1]]))),elem[1])
           # print(newRay)
           # print("initRay",newRay)
            if(0<=newRay[0][0]<maxX and 0<=newRay[0][1]<maxY):
                case= mapp[newRay[0][0]][newRay[0][1]]
         #       print(case)
                if(case in ["/","\\"]):
                    #print("mirror")
                    newRay=mirror(newRay,case)
                    #print("newRay",newRay)
                 #   if(not newRay in newList):
                    newList.append(newRay)
                    # seen.add(newRay)
                elif(case in ["-","|"]):
                    #print("splitter")
                    newRays=splitter(newRay,case)
                    #print("newRay",newRay)
                    for ray in newRays:
                    #    if(not newRay in newList):
                        newList.append(ray)
                        # seen.add(ray)
                else:
                 #   if(not newRay in newList):
            #        print("new")
                    newList.append(newRay)
                  #  seen.add(newRay)

        for elem in newList:
            energized[elem[0]]=True
        newEner=len(energized.keys())
        # if(not newEner>previousEner):
        #     count+=1
        #     if(count>=50):
        #         break
        previousEner=newEner
        rayList=newList
     #   print("raylist",rayList)
        
       # print("rayList",rayList)
    return rayList,energized
            
        
rayListSilver=[]
initRaySilver=((0,0),3)

rayListSilver.append(initRaySilver)
energizedSilver={}
maxX=len(inp)
maxY=len(inp[0])
rayListSilver,energizedSilver=part1(energizedSilver,rayListSilver,dirr,inp)
print("silver:",len(energizedSilver.keys()))
energieList=[]
for i in range(len(inp)):        
    for j in range(0,4):
        rayList=[]
        initRay=((0,i),j)
     #   print(initRay)
        rayList.append(initRay)
        energized={}
        rayList,energized=part1(energized,rayList,dirr,inp)
        energieList.append([len(energized.keys())])
     #   print(len(energized.keys()))

        rayList=[]
        initRay=((maxX-1,i),j)
        rayList.append(initRay)
        energized={}
      #  print(initRay)
        rayList,energized=part1(energized,rayList,dirr,inp)
        energieList.append([len(energized.keys())])
     #   print(len(energized.keys()))

for i in range(maxY):
    for j in range(0,4):
        rayList=[]
        initRay=((i,0),j)
        rayList.append(initRay)
        energized={}
        rayList,energized=part1(energized,rayList,dirr,inp)
        energieList.append([len(energized.keys())])
      #  print(len(energized.keys()))


        rayList=[]
        initRay=((i,maxY-1),j)
        rayList.append(initRay)
        energized={}
     #   print(initRay)
        rayList,energized=part1(energized,rayList,dirr,inp)
        energieList.append([len(energized.keys())])
     #   print(len(energized.keys()))
print("Gold",max(energieList))
```
