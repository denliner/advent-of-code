```python
import re,cProfile
from collections import defaultdict
from functools import cache
reg="-?[0-9]+"
INPUT="input.txt"
#print(re.findall(reg,"1,2,3"))
springs=defaultdict(dict)
springsP2=defaultdict(dict)
with open(INPUT) as f:
    inp=f.read().strip().splitlines()
    inp= [x.split(" ") for x in inp]

for i,elem in enumerate(inp):
    springs[i]["map"]=elem[0]
    #print(elem)
    springs[i]["damaged"]=list(map(int,re.findall(reg,elem[1]) )) 

    springsP2[i]["map"]=""+springs[i]["map"]
    #print(elem)
    springsP2[i]["damaged"]=[]+springs[i]["damaged"]
    for j in range(4):
        springsP2[i]["map"]= springsP2[i]["map"]+"?" + springs[i]["map"]
        springsP2[i]["damaged"]= springsP2[i]["damaged"] + springs[i]["damaged"]



def DFS(mapp,damaged,visited):
    if(mapp in visited):
        return
    for i in range(len(mapp)):
        if(mapp[i]=='?'):
            copyMapp=list(mapp)
            copyMapp[i]='#'
            cope="".join(copyMapp)
            DFS(cope,damaged,visited)
            visited.add(cope)
    visited.add(mapp)
@cache
def countCombi(mapp,damaged,currentSize):
    lenDamaged=len(damaged)
    summ=0

    
    if(len(mapp)==0):
        if(lenDamaged==0 and currentSize==0):
            return 1
        elif(lenDamaged==1 and currentSize==damaged[0]):
            return 1
        else:
            return 0
   

    if(lenDamaged>0 and currentSize>damaged[0]):
        return 0
    elif(lenDamaged==0 and currentSize>0):
        return 0

    start=mapp[0]
    if(start in ["#","?"] ):
        summ+=countCombi(mapp[1:],damaged,currentSize+1)
    if(start in [".","?"] ):
        if(lenDamaged>0 and damaged[0] ==currentSize):
            summ+=countCombi(mapp[1:],damaged[1:],0)
        elif(currentSize==0):
            summ+=countCombi(mapp[1:],damaged,0)
    return summ

    
def countDamaged(sp,j):
    vis=set()
    DFS(sp[j]["map"],sp[j]["damaged"],vis)
    ListVis=list(vis)
    lenV=len(ListVis)
    for i in range(lenV):
        ListVis[i]=ListVis[i].replace("?",".")
        ListVis[i]=ListVis[i].replace("."," ").split()
    
    sizeVis= [ [len(x) for x in y] for y in ListVis ]
    return sizeVis.count(sp[j]["damaged"])
def part1(sp):
    summ=0
    ran=len(sp.keys())
    print("count")
    for i in range(ran):
        summ+=countDamaged(sp,i)
    print(summ)
    return summ

def part2(sp):
    summ=0
    for i in springs:
        summ+=countCombi(tuple(sp[i]["map"]),tuple(sp[i]["damaged"]),0)
    print(summ)
    return summ
    

#cProfile.run('part1(springs)')

# def part2(spring):
#     summ=0
#     print("count")
#     for i in range(len(springs.keys())):
#         summ+=countDamaged(springs,i)
#     print(summ)
#     return summ

cProfile.run('part1(springs)')
cProfile.run('part2(springsP2)')
```
