```python
INPUT="input.txt"
from copy import deepcopy
import cProfile
almanac={}
almanacP2={}
with open(INPUT) as f:
    inp=f.read().strip().split("\n\n")
    inp=[x.replace(": ",":\n").split(":\n") for x in inp]
    inp=[[x[0],x[1].split("\n")] for x in inp]

for i,elem in enumerate(inp):
    for j,x in enumerate(inp[i][1]):
        inp[i][1][j]= list(map(int,inp[i][1][j].split(" ")))

for elem in inp:
    if(elem[0]=="seeds"):
        almanac[elem[0]]=elem[1][0]
        almanacP2[elem[0]]=[]
    else:
        rangeList=[]
        rangeList2=[]
        for x in elem[1]:
            r=range(x[1],x[1]+x[2])
            r2=range(x[0],x[0]+x[2])
            destination=x[0]-x[1]
            destination2=x[1]-x[0]

            rangeList.append([r,destination])
            rangeList2.append([r2,destination2])
        almanac[elem[0]]=rangeList
        almanacP2[elem[0]]=rangeList2
        
finalSeeds=[]
for seed in almanac['seeds']:
    s=seed
  #  print("ini",s)
    for elem in almanac:
        if(elem=="seeds"):
            continue
        mapp=almanac[elem]
        for m in mapp:
            if(s in m[0]):
                s+=m[1]
        #        print(s,m,elem)
                break
    finalSeeds.append(s)
   # print("final:",s,"\n")

# almanacP2=deepcopy(almanac)
#for x,y in almanacP2["seeds"]:
#    print(x,y)

listDuo=[]
for x, y in zip(almanac["seeds"][::2], almanac["seeds"][1::2]):
    listDuo.append(range(x,x+y+1))
#min(finalSeeds)
almanacP2["seeds"]=listDuo


def reverseS(almanacP2):
    keyP2=list(reversed(almanacP2.keys()))
    i=0
    while True:
      #  print("ini",s)
        s=i
        for elem in keyP2:
            if(elem=="seeds"):
                for ran in almanacP2["seeds"]:
                    if(s in ran):
                        return i
            else:
                mapp=almanacP2[elem]
                for m in mapp:
          #          print("m",m)
                    if(s in m[0]):
                        s+=m[1]
                #        print(s,m,elem)
                        break
        i+=1
       # print("final:",s,"\n")


# finalSeedsP2=[]
# for rang in almanacP2['seeds']:
#     for seed in rang:
#         s=seed
#       #  print("ini",s)
#         for elem in almanacP2:
#             if(elem=="seeds"):
#                 continue
#             mapp=almanacP2[elem]
#             for m in mapp:
#                 if(s in m[0]):
#                     s+=m[1]
#             #        print(s,m,elem)
#                     break
#         finalSeedsP2.append(s)
#        # print("final:",s,"\n")
almanacP2,
#min(finalSeeds),
cProfile.run("reverseS(almanacP2)") 
```
