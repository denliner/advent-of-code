```python
INPUT="input.txt"
from collections import defaultdict
import re

cardList=defaultdict(dict)

with open(INPUT) as f:
    inp=f.read().strip().splitlines()
    inp=[x.split(": ")[1] for x in inp  ]
    inp=[x.split(" | ") for x in inp  ]

score=0
for i,elem in enumerate(inp):
    winning=list(map(int,re.findall("[0-9]+",elem[0]))) 
    own=list(map(int,re.findall("[0-9]+",elem[1]))) 
    inter=list(set(winning) & set(own))
    cardScore=0
    interNum=len(inter)
    if(inter):
        cardScore=pow(2,interNum-1)
        score+=cardScore
    cardList[i+1]["score"]=cardScore
    cardList[i+1]["number"]=1
    cardList[i+1]["interNum"]=interNum
score,cardList
scoreP2=0
for key in cardList:
    if(cardList[key]["score"]>0):
        for j in range(1,cardList[key]["interNum"]+1):
            if(key+j in cardList.keys()):
                cardList[key+j]["number"]+=1*cardList[key]["number"]

for key in cardList:
    scoreP2+=cardList[key]["number"]
score,scoreP2,cardList
```
