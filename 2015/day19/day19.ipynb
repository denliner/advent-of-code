```python
INP="input.txt"
reg="([A-Z])(?![A-Z])"
import re
from collections import defaultdict
with open(INP) as f:
    impD=f.read().strip().split("\n\n")
    imp=impD[0].splitlines()
    word=re.findall("[A-Z][a-z]*",impD[1])


molecule=defaultdict(list)
wordList=[]
for elem in imp:
    sp=elem.split(" => ")
    molecule[sp[0]].append(sp[1])
molecule,word
for elem in molecule:
    molecule[elem].append(elem)
def DFS(word,strin):
    if(word):
        first=word[0]
        for elem in molecule[first]:
            st=strin+elem
            DFS(word[1:],st)
            
    else:
        wordList.append(strin)
        return True

def DFS1(word,strin,isDone):
    if(word):
        first=word[0]
       # print(first)
        if(not isDone):
            if(first in molecule):
                for elem in molecule[first]:
                    st=strin+elem
                  #  print(st)
                    done= False if (elem==first) else True
                    DFS1(word[1:],st,done)
            else:
                st=strin+first
               # print(st)
                DFS1(word[1:],st,False)
        else:
            DFS1(word[1:],strin+word[0],True)
            
    else:
        wordList.append(strin)
        return True
DFS1(word,"",False)
final=list(set(wordList))
final= [x for x in final if (not x==impD[1]) ]
len(final),wordList,molecule
```
