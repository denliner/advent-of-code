```python
INPUT="input.txt"
import re

with open(INPUT) as f:
    imp= list(map(int,re.findall("[-]?[0-9]+",f.read())))
    
class Node:
    def __init__(self, v=None):
        self.v = v
        self.n = None
        self.p = None
        
def generate_Nodes(liste,bias):
    dictionary = {}
    zero=None
    start,last = None, None
    for i,element in enumerate(liste):
        node = Node(element*bias)
        if last is not None:
            last.n = node
            node.p = last
        else:
            start = node
        dictionary[i] = node
        last = node
        if(element==0):
            zero=node
    last.n = start
    start.p = last
    return dictionary,zero


def mixing(dictionary,maxx, times):
    for _ in range(times):
        for i in range(maxx):
            current = dictionary[i]
            number = dictionary[i].v%(maxx-1)
            if(not number == 0):
                first = current
                if(number > 0):
                    for j in range(number):
                        first = first.n
                else:
                    for j in range((abs(number)+1)):
                        first = first.p


                second = first.n
                beforeC = current.p
                afterC = current.n

                first.n = current
                current.n = second
                current.p = first
                second.p = current
                beforeC.n = afterC
                afterC.p = beforeC
    return dictionary
    
def day20(imp,bias,times):
    maxx=len(imp)
    dictionary,zero = generate_Nodes(imp,bias)
    dictionary = mixing(dictionary,maxx,times)
    index=zero
    result = 0
    for i in range(3001):
        if(i in [1000,2000,3000] ):
            result+= index.v
        index = index.n
    return result

%time print("silver: ",day20(imp,1,1)),
%time print("gold: ",day20(imp,811589153,10))
```
