```python
INPUT="input.txt"
with open(INPUT) as f:
    imp=f.read().splitlines()
    imp=[list(map(int,list(x))) for x in imp]
    
def check(imp,indice,direction,height,width):
    sign=1
    stepI, stepJ=0,0
    match direction:
        case "u":
            sign=-1
            stepI=indice[0]+1
        case "d":
            sign=1
            stepI=height-indice[0]
        case 'l':
            sign=-1
            stepJ=indice[1]+1
        case 'r':
            sign=1
            stepJ=width-indice[1]
    for i in range(1,stepI):
        neigPosition= indice[0]+i*sign,indice[1]
        #print(indice[0]+i*sign,indice[1],neigPosition)
        if(imp[neigPosition[0]][neigPosition[1]]>= imp[indice[0]][indice[1]] ):
            return False
    #print("LR")
    for j in range(1,stepJ):
        neigPosition= indice[0],indice[1]+j*sign
        #print(indice[0],indice[1]+j*sign,neigPosition)
        if(imp[neigPosition[0]][neigPosition[1]]>= imp[indice[0]][indice[1]] ):
            return False
    return True

def checkP2(imp,indice,direction,height,width):
    sign=1
    visible=True
    distance=0
    stepI, stepJ=0,0
    match direction:
        case "u":
            sign=-1
            stepI=indice[0]+1
        case "d":
            sign=1
            stepI=height-indice[0]
        case 'l':
            sign=-1
            stepJ=indice[1]+1
        case 'r':
            sign=1
            stepJ=width-indice[1]
    for i in range(1,stepI):
        distance=i
        neigPosition= indice[0]+i*sign,indice[1]
        #print(indice[0]+i*sign,indice[1],neigPosition)
        if(imp[neigPosition[0]][neigPosition[1]]>= imp[indice[0]][indice[1]] ):
            break
    for j in range(1,stepJ):
        distance=j
        neigPosition= indice[0],indice[1]+j*sign
        #print(indice[0],indice[1]+j*sign,neigPosition)
        if(imp[neigPosition[0]][neigPosition[1]]>= imp[indice[0]][indice[1]] ):
            break
    return distance
count=0
height,width = len(imp) , len(imp[0])
dirr=["u","d","l","r"]
for i,line in enumerate(imp):
    for j,elem in enumerate(line):
        if(i==0 or i==(height-1) 
           or j==0 or j==(width-1)):
            count+=1
        else:
            for d in dirr:
                if(check(imp,[i,j],d,height,width)):
                    count+=1
                    break
highestScore=0
for i,line in enumerate(imp):
    for j,elem in enumerate(line):
        score=1
        for d in dirr:
            score*= checkP2(imp,[i,j],d,height,width)
        if(score>highestScore):
            highestScore=score
test=1
for d in dirr:
    result=checkP2(imp,[3,2],d,height,width)
    test*= result
    print(d,test,result)
count,highestScore,test
```
