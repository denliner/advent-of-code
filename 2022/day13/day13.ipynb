```python
INPUT="input.txt"
import json

with open(INPUT) as f:
    imp=f.read().split("\n"*2)
    imp=[x.splitlines() for x in imp]
    imp= [ [eval(y) for y in x] for x in imp]

def compare(first,second):
   # print(first,second,"\n"*2)
    final=False
    if( (first is None) and (second is None) ):
        return True
    size1, size2 = len(first),len(second)
    size= size1 if size1>size2 else size2
    result=None
    for i in range(size1):
        try:
            second[i]
        except IndexError:
            return False         
        if (type(first[i]) is  list 
           and  type(second[i]) is  list):
            result=compare(first[i],second[i])
        elif(type(first[i]) is  list):
            result= compare(first[i],[second[i]])
        elif(type(second[i]) is  list):
            result= compare([first[i]],second[i])
        elif(type(first[i]) is  int 
           and  type(second[i]) is  int) :
            #print("not es list",first[i],second[i],type(first[i]))
            if(first[i]==second[i]):
                continue
            elif(first[i]<second[i]):
             #   print(leftL[i],rightL[i],True)
                return True
            else:
              #  print(leftL[i],rightL[i],"",False)
                return False
        if(not result is None):
            return result
    if(size1<size2):
        return True
    return None

def merge(left,right):
    if not left:
        return right
    elif not right:
        return left
    elif compare(left,right):
        return [left[0]] + merge(left[1:],right)
    else:
        return [right[0]] + merge(left,right[1:])

def mergeSort(tab):
    length=len(tab)
    if length<=1:
        return tab
    else:
        return merge(mergeSort(tab[:length//2]), mergeSort(tab[length//2:]))
def day13(imp):

    c=0
    apCouple=[]
    #print(imp[0][0])
    for i,tup in enumerate(imp):
      #  print(len(tup))
        if(len(tup)<1):
            continue
        left, right =  tup[0],tup[1]
        check = compare(left,right)
        apCouple.append(tup[0])
        apCouple.append(tup[1])
        if(check):
            #print("INDICE",i+1)
          #  apCouple.append(i+1)
            c+=i+1
    #print(ap)
    apCouple
    apCouple.append([[6]])
    apCouple.append([[2]])
    newList=mergeSort(apCouple)
    return c,(newList.index([[6]])+1)*(newList.index([[2]])+1)
    #apCouple
#print(imp[0][0])
%time print(day13(imp))
```
