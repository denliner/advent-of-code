```python
INPUT="input.txt"

class Directory:
    def __init__(self, parentDir=None, name=""):
        self.subDir = []
        self.files= []
        self.parentDir = parentDir
        self.size= None
        self.name = name
    def getSize(self):
        if(not self.size):
            #print(self.name,"size")
            s=0
            for f in self.files:
                #print(f.name,self.name,f.size)
                s+=f.size
            for dirr in self.subDir:
                #print("dir",dirr.name,self.name,f.size)
                s+=dirr.getSize()
            self.size=s
            return s
        else:
            return self.size

class file:
    def __init__(self, size=None, name=None):
        self.size = size
        self.name = name
    

def Spaghetti(listNamedDir,listDir,name,currentDir):
    indices = [i for i, x in enumerate(listNamedDir) if x == name]
    #print(indices)
    for i in indices:
        if(listDir[i].parentDir==currentDir):
            return True
    return False

    
def getDir(listCommand):
    currentDir=None
    listDir=[]
    listNamedDir=[]
    for i,commande in enumerate(listCommand):
        #print(commande,len(listCommand))
        if("cd" in commande[0]):
            if(commande[0][1]==".."):
                #print(commande,i)
                if(currentDir.parentDir):
                    dirr=currentDir.parentDir
            # elif((not commande[0][1] in listNamedDir) or 
            #       not Spaghetti(listNamedDir,listDir,commande[0][1],currentDir) ):
            else:
                # print(commande[0][1])
                dirr=Directory(parentDir=currentDir,name=commande[0][1])
                if(currentDir):
                    currentDir.subDir.append(dirr)
                listDir.append(dirr)
                listNamedDir.append(dirr.name)
            # else:
            #     # print("spag")
            #     dirr=listDir[listNamedDir.index(commande[0][1])]
            currentDir=dirr
            
        elif("ls" in commande[0]):
            #print("ls")
            #print(commande[0])
            for elem in commande[1:]:
                typesorSize= elem[0]
                name= elem[1]
                #print("typesorSize",typesorSize)
                if typesorSize=="dir":
                    #print("dir")
                    dire=None
                    if(name in listNamedDir and Spaghetti(listNamedDir,listDir,name,currentDir) ):
                        dire=listDir[listNamedDir.index(name)]
                    else:
                      #  print(elem)
                        dire=Directory(parentDir=currentDir,name=elem[1])
                        listDir.append(dire)
                        listNamedDir.append(dire.name)
                    currentDir.subDir.append(dire)
                else:
                    #print("file")
                    newFile=file(size=int(typesorSize),name=name)
                    currentDir.files.append(newFile)
                
                    
    return listDir,listNamedDir
            
def day07():
    with open(INPUT) as f:
        imp=f.read().split("$")
        imp= list(filter(None, imp))
        imp = [x.strip() for x in imp]
        imp= [ x.split("\n") for x in imp ]
        imp= [ [y.split(" ") for y in x] for x in imp ]

    currentDir=None
    #print(imp[1][1:][0])
    listDir,listNamedDir = getDir(imp)
    silver=0
    goldList,sizeList=[], []
    for dirr in listDir:
        #print("check:",dirr.name)
        size=dirr.getSize()
        # print("\n",dirr.name,size,"\n")
        if size<=100000:
            silver+=size

        if(size>0):
            sizeList.append(size)
    unusedSize=70000000 - max(sizeList)
    goldList=[x for x in sizeList if  unusedSize + x > 30000000]

    #print(listNamedDir)
    # for dirr in listDir:
    #     print("check:",dirr.name)
    #     print("dir")
    #     for elem in dirr.subDir:
    #         print(elem.name)
    #     for elem in dirr.files:
    #         print(elem.name)
    #     print("\n"*2)
    return silver,min(goldList)
%time day07()
```
