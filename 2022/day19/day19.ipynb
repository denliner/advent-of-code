```python
import regex as re
from functools import cache
INPUT="input.txt"

with open(INPUT) as f:
    imp=f.read().strip().splitlines()
    imp=[x.replace("obsidian.","obpsidian").replace(":",".").split(". ")[1:] for x in imp]


def defBlueTuple(bp):
    need=()
    for i,elem in enumerate(bp):
        cost=list(map(int,re.findall("\d+",elem)))
        #print(elem)
        if(len(cost)>1):
            need= need + (tuple(cost),)
        else:
            need= need + (cost[0],)
    return need

@cache
def bruteforceDFSCached(route,robots,ores,costs,oreMax):
    maximum=0
    if(route<1):
        return ores[-1]
    condOre= (robots[0] < oreMax 
              and costs[0] <= ores[0] 
              and route>(costs[0]) )
    
    condObsidian = (robots[2] <  costs[3][1] 
                    and  costs[2][0] <= ores[0]
                    and costs[2][1] <= ores[1] and route>3 )
    
    condClay =   (robots[2] <  costs[3][1] 
                  and robots[1] < costs[2][1]
                  and costs[1]<= ores[0] and route>5)
    
    condGeode = (costs[3][0] <= ores[0] 
                 and costs[3][1]<= ores[2])
    
    if(condGeode):
        newRobot = tuple([robots[0],robots[1],robots[2],robots[3]+1])
        newOres  = tuple([ores[0]+robots[0]-costs[3][0],ores[1]+robots[1],ores[2]+robots[2]-costs[3][1],ores[3]+robots[3]])
        maximum = max(maximum,bruteforceDFSCached(route-1,newRobot,newOres,costs,oreMax))  
        
    else:
        newRobot = tuple([robots[0],robots[1],robots[2],robots[3]])
        newOres  = tuple([ores[0]+robots[0],ores[1]+robots[1],ores[2]+robots[2],ores[3]+robots[3]])
        maximum = max(maximum,bruteforceDFSCached(route-1,newRobot,newOres,costs,oreMax))     
        if(route>1):
            if(condObsidian):
                newRobot = tuple([robots[0],robots[1],robots[2]+1,robots[3]])
                newOres  = tuple([ores[0]+robots[0]-costs[2][0],ores[1]+robots[1]-costs[2][1],ores[2]+robots[2],ores[3]+robots[3]])
                maximum = max(maximum,bruteforceDFSCached(route-1,newRobot,newOres,costs,oreMax))   
            else:
                if(condClay):
                        newRobot = tuple([robots[0],robots[1]+1,robots[2],robots[3]])
                        newOres  = tuple([ores[0]+robots[0]-costs[1],ores[1]+robots[1],ores[2]+robots[2],ores[3]+robots[3]])
                        maximum =max(maximum,bruteforceDFSCached(route-1,newRobot,newOres,costs,oreMax))


                if(condOre):
                    newRobot = tuple([robots[0]+1,robots[1],robots[2],robots[3]])
                    newOres  = tuple([ores[0]+robots[0]-costs[0],ores[1]+robots[1],ores[2]+robots[2],ores[3]+robots[3]])
                    maximum =max(maximum,bruteforceDFSCached(route-1,newRobot,newOres,costs,oreMax))


                if(condOre):
                    newRobot = tuple([robots[0]+1,robots[1],robots[2],robots[3]])
                    newOres  = tuple([ores[0]+robots[0]-costs[0],ores[1]+robots[1],ores[2]+robots[2],ores[3]+robots[3]])
                    maximum =max(maximum,bruteforceDFSCached(route-1,newRobot,newOres,costs,oreMax))


    
    return maximum


silver,gold=0,1

for i,elem in enumerate(imp):
    costs=defBlueTuple(elem)
    maxx= max(costs[0],costs[1],costs[2][0],costs[3][0] )
    ad=bruteforceDFSCached(24,(1,0,0,0),(0,0,0,0),costs,maxx)
    silver+= ad*(i+1)
    bruteforceDFSCached.cache_clear()
    if(i<3):
        gold*=bruteforceDFSCached(32,(1,0,0,0),(0,0,0,0),costs,maxx)
        bruteforceDFSCached.cache_clear()


print(silver,gold)
```
