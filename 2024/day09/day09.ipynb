```python
import re,time
from copy import deepcopy
st= time.time()
INPUT="input.txt"


def getBlockSize(nS):
    lNBlock={}
    lDBlock=[]
    currentBlock=None
    firstId=0
    currentSize=0
    for i in range(len(nS)):
        if(nS[i]=="."):
            if((not currentBlock) or currentBlock=="."):
                currentBlock="."
                currentSize+=1
            else:
                new=[firstId,currentSize]
                firstId=i
                lNBlock[int(currentBlock)]=new
                currentBlock="."
                currentSize=1
        else:
            if((not currentBlock) or (currentBlock==nS[i])):
                currentBlock=nS[i]
                currentSize+=1
            else:
                new=[firstId,currentSize]
                if(currentBlock=="."):
                    lDBlock.append([firstId,currentBlock,currentSize])
                else:
                    lNBlock[int(currentBlock)]=new
                firstId=i
                currentBlock=nS[i]
                currentSize=1  
    new=[firstId,currentSize]
    if(currentBlock=="."):
        lDBlock.append([firstId,currentBlock,currentSize])
    else:
        lNBlock[int(currentBlock)]=new
    return lNBlock,lDBlock




with open(INPUT) as f:
    inp=f.read().strip()
newS=[]
nextNum=0
for i,num in enumerate(inp):
    count=int(num)
    if(True):
        if(i%2==0):
            nE=str(nextNum)
            nextNum+=1
        else:
            nE="."
        for c in range(count):
            newS.append(nE)

dotInd = [x for x in range(len(newS)) if newS[x] == "."]
numInd = [x for x in range(len(newS)) if not newS[x] == "."]

newS,dotInd,numInd
newS2=deepcopy(newS)

#print("".join(newS))
newS=list(newS)
while True:
    if(min(dotInd)>max(numInd)):
        break
    newS[dotInd[0]],newS[numInd[-1]] = newS[numInd[-1]],newS[dotInd[0]]
    numInd[-1] =dotInd[0]
    dotInd.pop(0)
    numInd.sort()


checksum,checksum2=0,0
lInt=[int(x) for x in newS if not x=="."]
for i,number in enumerate(lInt):
    checksum+=i*number
    
blockNum,blockDot=getBlockSize(newS2)
#print("".join(newS2))
maxKey=max(blockNum.keys())
bias=0
for i in range(maxKey,-1,-1):
    if(i not in blockNum.keys()):
        continue
   # print(i,blockNum)
    lenN=len(str(i))
    checkLen=blockNum[i][1]

    for j,dots in enumerate(blockDot):
        if(blockDot[j][0]>=blockNum[i][0]):
            continue
        if(blockDot[j][2]>=checkLen):
            for k in range(checkLen):

                newS2[blockDot[j][0]+k]=str(i)
                newS2[blockNum[i][0]+k]="."

            blockNum,blockDot=getBlockSize(newS2)  
            #print("".join(newS2))
            break

newS
#print("".join(newS2))

lInt2=[int(x) for x in newS2 if not x=="."]

for i,number in enumerate(newS2):
    if(number=="."):
        continue
    checksum2+=i*int(number)
et= time.time()
finalTime= et-st
print("time: "+str(finalTime))
checksum,checksum2
```
